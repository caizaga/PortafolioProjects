```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
with open('Alldef.txt', 'r') as file:
    text = file.read()
```

```python
text = text.lower()
text = text.replace(',', '').replace('.', '')
```

```python
nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
sentiment_scores = sid.polarity_scores(text)
```

```python
positive_words = []
negative_words = []
neutral_words = []
```

```python
for word in text.split():
    if sid.polarity_scores(word)['compound'] >= 0.5:
        positive_words.append(word)
    elif sid.polarity_scores(word)['compound'] <= -0.5:
        negative_words.append(word)
    else:
        neutral_words.append(word)
```

```python
positive_wordcloud = WordCloud(max_words=1000, width=800, height=400, background_color='white').generate(' '.join(positive_words))
```

```python
negative_wordcloud = WordCloud(max_words=1000, width=800, height=400, background_color='white').generate(' '.join(negative_words))
```

Positive and Negative Wordclouds

```python
plt.figure(figsize=(20, 10))
plt.subplot(1, 2, 1)
plt.imshow(positive_wordcloud, interpolation='bilinear')
plt.title('Positive Words')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(negative_wordcloud, interpolation='bilinear')
plt.title('Negative Words')
plt.axis('off')

plt.show()
```
